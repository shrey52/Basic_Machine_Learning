```python
#Importing the needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Importing the comma seperated value file 
data = pd.read_csv("4_Housing.csv")
```

```python
data
data.head(7)       #prints the first n rows in the dataset 
```

```python
print(data.keys())         #prints all the variables 
```

```python
data.shape                 #gives the rows and columns of the array
```

```python
#extracting the rows from the dataset 
X = data.iloc[:,0].values.reshape(-1,1)        #independent variable
Y = data.iloc[:,1].values.reshape(-1,1)        #dependent variable
```

```python
#Creating the model, fitting the values in the model and predicting some values
linear_regressor =LinearRegression()
linear_regressor.fit(X,Y)
Y_pred = linear_regressor.predict(X)
```

```python
Y_pred
```

```python
#Scatter plot for better understanding
plt.scatter(X,Y)
plt.plot(X,Y_pred,color='red')
plt.show()
```
